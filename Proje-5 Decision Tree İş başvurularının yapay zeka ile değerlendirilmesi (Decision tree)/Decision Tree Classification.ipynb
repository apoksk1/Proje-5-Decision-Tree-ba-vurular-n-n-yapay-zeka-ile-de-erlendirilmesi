```python
import numpy as no 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import tree

df = pd.read_csv("DecisionTreesClassificationDataSet.csv")
```

```python
df.head()
```

scikit-learn ktphanesi decision tree'lerin dzgn almas iin hereyin rakamlsal olmasn bekliyor bu nedenle veri setimizdeki tm Y ve N deerlerini 0 ve 1 olarak dzeltiyoruz. Ayn sebeple eitim seviyesini de BS:0 MS:1 ve PhD:2 olarak gncelliyoruz. map() kullanarak bo hcreler veya geersiz deer girilen hcreler NaN ile doldurulacaktr, buna uandaki veri setimizde ihtiyackz yok ama sizin ilerde youn veri ile altnz zaman ihtiyacnz olacaktr.

```python
duzeltme_mapping = {'Y':1, 'N': 0}

df['IseAlindi'] = df['IseAlindi'].map(duzeltme_mapping)
df['SuanCalisiyor?'] = df['SuanCalisiyor?'].map(duzeltme_mapping)
df['Top10 Universite?'] = df['Top10 Universite?'].map(duzeltme_mapping)
df['StajBizdeYaptimi?'] = df['StajBizdeYaptimi?'].map(duzeltme_mapping)

duzeltme_mapping_eitim ={'BS':0,'MS':1,'PhD':2}
df['Egitim Seviyesi'] = df['Egitim Seviyesi'].map(duzeltme_mapping_eitim)

df.head()
```

```python
y = df.iloc[:,-1].values

X = df.drop(['IseAlindi'], axis=1)
```

```python
clf =tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
# 5 yl deneyim bir yerde alan ve 3 eski irkette alm olan eitim seviyesi lisan
#top-tier-school mezunu deil
print(clf.predict([[5,1,3,0,0,0]]))
```
